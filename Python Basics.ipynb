```python
# Q4. Write a Python program to take a user’s name as input and print a greeting message.

name = input("Enter your name: ")
print("Hello " + name + "!, Welcome to Python")
```

```python
# Q5. Write a Python program to check whether a number is even or odd using conditional statements.

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
# Q6. Write a program to print numbers from 1 to 10 using a loop.

for i in range(1, 11):
    print(i)
```

```python
# Q7. Create a list of five numbers and print the maximum number from the list.

numbers = [10, 25, 5, 40, 15]

print("Maximum number is:", max(numbers))
```

```python
# Q8. Write a Python program to remove duplicate values from a list using a set.

numbers = [1, 2, 2, 3, 4, 4, 5]

unique_numbers = list(set(numbers))

print(unique_numbers)
```

```python
# Q9. Write a function that returns the square of a number.

def square(num):
    return num * num
```

```python
# Q10. Write a Python program to count how many times a number appears in a list.
# Example list [2,3,4,2,5,2]

numbers = [2, 3, 4, 2, 5, 2]

num = 2

count = numbers.count(num)

print("The number", num, "appears", count, "times in the list")
```
