# Libraries: Numpy & Pandas Assignment

```python
# Import Libraries

import pandas as pd
import numpy as np
```

```python
# Q1. Load the given retail dataset into a Pandas DataFrame and display the first 5 rows along with the dataset information to understand its structure.

# Load dataset
df = pd.read_csv("C:\\Users\\91829\\Downloads\\retail_dataset .csv") 

# Display first 5 rows
print(df.head())

# Dataset info
print(df.info())
```

```python
# Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing values present in the dataset.

# Missing values in each column
print(df.isnull().sum())

# Total missing values in dataset
print(df.isnull().sum().sum())
```

```python
# Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.

df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())
```

```python
# Q4. Remove all rows where the Product Category or Region column contains missing values.

df.dropna(subset=['Product Category', 'Region'], inplace=True)
```

```python
# Q5. Create a new column called Revenue using NumPy, where:
# Revenue = Qunatity * Price
# Display the updated dataset with the new column.

df['Revenue'] = df['Quantity'] * df['Price']

# Display updated dataset
print(df.head())
```

```python
# Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.

total_revenue = np.sum(df['Revenue'])

print("Total Revenue:", total_revenue)
```

```python
# Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.

category_revenue = df.groupby('Product Category')['Revenue'].sum()

print(category_revenue)
```

```python
# Q8. Based on the revenue calculated for each product category, identify top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.

# Top 3
top_3 = category_revenue.sort_values(ascending=False).head(3)

# Bottom 3
bottom_3 = category_revenue.sort_values().head(3)

print("Top 3 Categories:\n", top_3)
print("Bottom 3 Categories:\n", bottom_3)
```

```python
# Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the region with the highest revenue and the region with the lowesr revenue.

region_revenue = df.groupby('Region')['Revenue'].sum()

# Highest & Lowest
highest_region = region_revenue.idxmax()
lowest_region = region_revenue.idxmin()

print(region_revenue)
print("Highest Revenue Region:", highest_region)
print("Lowest Revenue Region:", lowest_region)
```

```python
# Q10. Extract the month from the Data column using Pandas and group the dataset by month to calculate the total monthly revenue. 
# Also calculate the mean, median, and standard deviation of the Revenue column using NumPy.

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Extract month
df['Month'] = df['Date'].dt.month

# Monthly revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()
print(monthly_revenue)

# Statistics
mean_rev = df['Revenue'].mean()
median_rev = df['Revenue'].median()
std_rev = df['Revenue'].std()

print("Mean:", mean_rev)
print("Median:", median_rev)
print("Standard Deviation:", std_rev)
```
